# test_graph

用于直接验证 TaskRouterGraph 全链路。


```python
from pathlib import Path
import os
import sys

PROJECT_ROOT = next(
    candidate
    for candidate in [Path.cwd().resolve(), *Path.cwd().resolve().parents]
    if (candidate / 'src' / 'task_router_graph').exists()
)
SRC_ROOT = PROJECT_ROOT / 'src'
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
if str(SRC_ROOT) not in sys.path:
    sys.path.insert(0, str(SRC_ROOT))

from task_router_graph.graph import TaskRouterGraph

print('模块加载完成')
print('当前 Python：', sys.executable)
print('已检测 API_KEY_Qwen：', bool(os.getenv('API_KEY_Qwen')))
```

```python
graph = TaskRouterGraph(config_path='configs/graph.yaml')
result = graph.run_case(PROJECT_ROOT / 'data' / 'cases' / 'case_01.json')

output = result['output']
env_payload = result['environment']
rounds = env_payload['rounds']

# 结构断言：round 必须存在，且 round_id 递增
assert isinstance(rounds, list) and len(rounds) >= 1
assert [r['round_id'] for r in rounds] == list(range(1, len(rounds) + 1))

last_round = rounds[-1]
assert isinstance(last_round['tasks'], list) and len(last_round['tasks']) >= 1

# task_id 在 round 内递增
assert [t['task_id'] for t in last_round['tasks']] == list(range(1, len(last_round['tasks']) + 1))

last_task = last_round['tasks'][-1]
trace = last_task['controller_trace']
assert isinstance(trace, list) and len(trace) >= 1

# 当前 round（供观察）可由最后一个 round_id 直接得到
cur_round = last_round['round_id']

print('全链路执行完成')
print('任务类型：', output['task_type'])
print('任务状态：', output['task_status'])
print('round 数量：', len(rounds))
print('cur_round：', cur_round)
print('最后 round 的 task 数量：', len(last_round['tasks']))
print('最后 task 的动作条数：', len(trace))
print('最后动作：', trace[-1]['action_kind'])
```
